# Estimate of $\pi$

compare the area of 1/4 circle with a square

circumference: $x^2 + y^2 = r^2$

The $r=1$ so that $A_{\text{circ}} = \pi$

And square has $l=1$ so that $A_{\text{square}} = l^2 = 1$

$\rightarrow \frac{\pi}{4} : 1 = n_{\pi} : n_{\text{tot}}$
$\rightarrow \pi = 4 \frac{n_{\pi}}{n_{\text{tot}}}$

$\rightarrow \frac{\pi}{4} : 1 = n_{\pi} : n_{\text{tot}}$
$\rightarrow \pi = 4 \frac{n_{\pi}}{n_{\text{tot}}}$

```python
import sys
import matplotlib.pyplot as plt
import random

r = 1
l = 1

x = random.random()
y = random.random()
print(x,y)

circ = x*x + y*y
circ

def plot(N):
    
    x_grid,y_grid = pi_comp(1, N)[1]
    xc_grid,yc_grid = pi_comp(1, N)[2]
    
    plt.scatter(x_grid, y_grid, color='blue', s=10)
    plt.scatter(xc_grid, yc_grid, color='orange', s=5)
    plt.show()
    
    return

def pi_comp(r, N):
    
    counts_tot = []
    counts_pi = []
    x = []
    y = []
    x_c = []
    y_c = []
    
    for i in range(0, N):
        
        counts_tot.append(i)
        a = random.random()
        b = random.random()
        x.append(a)
        y.append(b)
        circ = a*a + b*b
        
        if circ <= r*r: 
            
            counts_pi.append(circ)
            x_c.append(a)
            y_c.append(b)
            
    pi_estimate = 4*len(counts_pi)/len(counts_tot)
        
    return pi_estimate, [x,y], [x_c, y_c]

def mean_err(N, n):
    
    pi_est = []
    
    for i in range(0,n):
        
        pi_est.append(pi_comp(1, N)[0])
        
    mean_pi = sum(pi_est)/len(pi_est)
    var = sum((i - mean_pi) ** 2 for i in pi_est) / len(pi_est)
        
    return mean_pi, var**(1/2)
            
```

### Result

```python
pi, err = mean_err(10000, 1000)
print("pi: ", pi, "\nerr: ", err)
```
